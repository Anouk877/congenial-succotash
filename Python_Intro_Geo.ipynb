## `IF`-Abfragen
Mit If-Abfragen kann man in Python **Bedingungen** programmieren. Grundsätzlich laufen diese Abfragen immer nach dem folgenden **Schema** ab:

```
If <Bedingung> wahr:
  <dann>
else:
  <sonst>
```

Als einfaches Beispiel programmieren wir uns hier eine If-Abfrage, welche für eine Summe von Zahlen prüft, ob sie größer oder kleiner als 100 ist.

````python
# Beispiel If-Abfrage
zaehlen = [10, 1, 69, 24]
summe = sum(zaehlen)

if summe > 100:
  print("Summe ist größer als 100")
else:
  print("Summe ist kleiner oder gleich 100")
````

Was das Programm also macht:
1.   Errechnen der Summer der Startzahlen
2.   Prüfen ob Summe größer als 100 ist:
3.   Wenn ja: „Summe größer als 100“
4.   Sonst: „Summe kleiner oder gleich 100“

Tatsächlich können in derselben if-Schleife noch weitere Fälle spezifiziert werden. Dies geschieht dann mit `elif`:

```
If <Bedingung> wahr:
  <dann>
elif <andere Bedingung> wahr:
  -dann-
(elif <weitere Bedingung> wahr:
  <dann>)
.
.
.
else:
  <sonst>
```

Dabei können auch mehrere „elifs“ hintereinander genutzt werden. Zum Schluss muss nur dann wieder „else“ verwendet werden.

## `while`-Schleifen
Bei der While-Schleife durchläuft die Schleife ein Teil eines Programms so lange, bis eine Bedingung nicht mehr zutrifft. Das Schema der While-Schleifen läuft folgendermaßen ab:

```
while <Bedingung> wahr:
  <Befehl ausführen>
```



````python
# Beispiel while-Schleife
i = 1
while i < 10:
  print(i)
  i = i + 1
````

Was das Programm also macht:
1.   Einen Variable i als „Laufindex“ definieren
2.   Solange i kleiner als 10 ist:
3.   i in der Konsole ausgeben
4.   i wird um 1 erhöht
Demnach wird die Schleife neun-mal durchlaufen bis der Laufindex i nicht mehr kleiner als 10 ist.

for <Element> in <z.B Liste>:
  <Befehl ausführen>

## Funktionen
Mit Funktionen kann ein Teilprogramm, welches öfter wiederholt wird, als eigenständige Funktion definiert werden. Diese kann dann immer wieder abgerufen werden. Die Definition von Funktionen läuft immer nach folgendem Schema ab:

```
def Funktionsname(Parameter_1, ..., Parameter_n):
  <Funktionsinhalt>
```



````python
# Beispiel Definition einer Funktion
def cum_sum(numbers):
  result = 0
  for num in numbers:
    result = result + num
  print(f"Ergebnis: {result}")

liste = [1, 23, 54, -54, 25]

cum_sum(liste)
````

Hier wurde eine Funktion definiert, welche für eine beliebige Liste von Zahlen die **kumulierte Summe** ausgibt.
Was das Programm also macht:
1.   Zuerst wird eine neue Funktion mit dem Namen `cum_sum` mit dem Parameter der zu berechnenden Liste definiert
2.   Dann werden zwei neue Variablen für das endliche Ergebnis und für einen Laufindex der Liste erstellt
3.   Nun wird jedes Element der Liste auf die Variable `result` addiert
4.   Schließlich wird das Ergebnis ausgegeben

## Aufgabe 3 — Schleifen & Funktionen
Implementiere die Haversine-Funktion, finde Station am weitesten von der ersten Station, und berechne `nearest_dist_km` für jede Station.

````python
import numpy as np
from math import radians, sin, cos, sqrt, atan2

def haversine_km(lon1, lat1, lon2, lat2):
    # Earth radius
    R = 6371.0
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2.0)**2 + cos(lat1) * cos(lat2) * sin(dlon/2.0)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c

# Entfernung zur ersten Station
lon0, lat0 = df.loc[0, ['lon','lat']]
distances_to_first = df.apply(lambda row: haversine_km(lon0, lat0, row['lon'], row['lat']), axis=1)
max_idx = distances_to_first.idxmax()
print(f"Station mit größter Entfernung zur ersten Station: ID={df.loc[max_idx,'id']}, Distanz={distances_to_first[max_idx]:.1f} km")


# Nächstliegende Nachbarn (naiver O(n^2) Schleifen-Lösung)
nearest = []
for i, row_i in df.iterrows():
    min_d = float('inf')
    for j, row_j in df.iterrows():
        if i == j:
            continue
        d = haversine_km(row_i['lon'], row_i['lat'], row_j['lon'], row_j['lat'])
        if d < min_d:
            min_d = d
    nearest.append(min_d)

df['nearest_dist_km'] = nearest
print('\nBeispiel: erste fünf nearest distances (km):')
display(df[['id','nearest_dist_km']].head())
````
